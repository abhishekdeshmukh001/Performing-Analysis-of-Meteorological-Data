```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('weatherHistory.csv')
```

```python
data['Formatted Date'] = pd.to_datetime(data['Formatted Date'], utc=True)
```

```python
data = data.set_index('Formatted Date')
cols = ['Apparent Temperature (C)','Humidity']
data_cols = data[cols].resample('MS').mean()
```

```python
data_cols
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.figure(figsize=(15,3))
plt.plot(data_cols['Humidity'], label='Humidity', color='green', linestyle='dashed')
plt.plot(data_cols['Apparent Temperature (C)'], label='Apparent Temperature', color='blue')
plt.title('Apparent Temperature Vs Humidity', fontsize=16)
plt.legend(fontsize=9)
plt.xticks(fontsize=13)
plt.yticks(fontsize=11)
```

```python
april = data_cols[data_cols.index.month==4]
```

```python
plt.figure(figsize=(15,3))
plt.plot(april.loc['2006-04-01':'2016-04-01', 'Apparent Temperature (C)'], linestyle='dashed', color='blue', marker='.', label='Apparent Temperature', alpha=0.8)
plt.plot(april.loc['2006-04-01':'2016-04-01', 'Humidity'], linestyle='dotted', color='purple', marker='o', label='Humidity')
plt.title('Apparent Temperature Vs Humidity', fontsize=16)
plt.legend(fontsize=9)
plt.xlabel('April Month', fontsize=14)
plt.xticks(fontsize=11)
plt.title('Humidity Vs Apparent Temperature', fontsize=17)
```

```python
import seaborn as sns
```

```python
plt.figure(figsize=(14,3))
sns.barplot(x='Apparent Temperature (C)', y='Humidity', data=april)
plt.xticks(rotation=45)
plt.xlabel('Apparent Temperature',fontsize=13)
plt.ylabel('Humidity',fontsize=12)
```
